# This notebook intends to take a quick look of profiling/summerizing done by pandas_profiling package.


```python
import pandas as pd
```

```python
from pandas_profiling import ProfileReport
```

```python
df=pd.read_csv("/Users/anurag/Desktop/Anurag_Koushal_Appinventiv_Submission/Assignment_2/Anurag_Koushal_Appinventiv/cleaned_data/cleaned_application_data.csv")
```

```python
profile = ProfileReport(
                        df,
                        title="Pandas Profiling Report",
                        explorative=True,
                        plot={"missing": {"cmap": "RdBu_r"}},
                        )
```

```python
profile.to_file(output_file="panda_profiling.html")
```
